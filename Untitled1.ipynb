```python
exam_score = []


for i in range(len(df['scores'])):


  exam_score.append(round(float(re.sub('^{.*: |}','',str(df['scores'][i][0]))),2))
```

```python
exam_score
```

```python
quiz_score=[]
for i in range(len(df['scores'])):
  quiz_score.append(round(float(re.sub('^{.*: |}','',str(df['scores'][i][1]))),2))
```

```python
quiz_score
```

```python
home_work_score=[]
for i in range(len(df['scores'])):
  home_work_score.append(round(sum(list(map(float,re.findall('\d+.\d+',str(df['scores'][i][2:])))))/len(list(map(float,re.findall('\d+.\d+',str(df['scores'][i][2:]))))),2))
```

```python
home_work_score
```
